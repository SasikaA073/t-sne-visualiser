## Load images

```python
# load images using 

import os 
image_file_names = os.listdir("images")

# sort images by name
image_file_names = sorted(image_file_names)
```

```python

image_file_paths = [os.path.join("images", image_file_name) for image_file_name in image_file_names]

image_file_paths[:5]
```

```python
# rename images to f"image_{i}.jpg"
for i, image_file_path in enumerate(image_file_paths):
    new_image_file_path = os.path.join("images", f"image_{i}.jpg")
    os.rename(image_file_path, new_image_file_path)
    image_file_paths[i] = new_image_file_path
```

```python
image_file_paths[:5]
```

```python
# show one image
from PIL import Image

image = Image.open(image_file_paths[0])

# reduce image size to 224x224 px
image = image.resize((224, 224))
```
